```python
import requests
from bs4 import BeautifulSoup
from keybert import KeyBERT
import collections
from urllib.parse import urljoin
```

```python
starting_url = "https://www.cc.gatech.edu/"

model = KeyBERT('distilbert-base-nli-mean-tokens')
```

```python
visited = set()
q = [starting_url]
storage = collections.defaultdict(list)

while len(q) > 0 and len(visited) < 1000:
    url = q.pop(0)
    if url not in visited:
        try: 
            x = requests.get(url, timeout=5)

            if x.status_code == 200:
                visited.add(url)
                print(len(visited))
                soup = BeautifulSoup(x.text, 'html.parser')
                relative_links = soup.find_all('a')
                text = soup.get_text(' ', strip=True)[:5000]
                keywords = model.extract_keywords(text)
                for kw in keywords:
                    storage[kw].append(url)

                for link in relative_links:
                    next_url = link.get("href")
                    if next_url and "cc.gatech.edu" in urljoin(url, next_url) and next_url not in visited:
                        q.append(urljoin(url, next_url))
        except:
            continue

print(len(visited))
        
```

### Crawl Statistics

```python
t = 1564 / float(60)
num_kw = len(storage)
print(num_kw)
num_crawled_url = 1000
num_extracted_url = num_crawled_url + len(q)

kw_p_s = float(num_kw) / t
crawled_url_p_s = float(num_crawled_url) / t
extracted_url_p_s = float(num_extracted_url) / t
crawled_extracted = float(num_crawled_url) / num_extracted_url

print(kw_p_s)
print(crawled_url_p_s)
print(extracted_url_p_s)
print(crawled_extracted)
```
